✅ Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400


```python
Room_charges_per_day = 2000
Doctor_conslt_fee = 1500
Lab_test_carge = 300

Stayed_days = 4
Lab_tests = 3
Medicine_bill = 2400

Room_cost = Room_charges_per_day * Stayed_days
Lab_cost = Lab_test_carge * Lab_tests

Total_Bill = Room_cost + Room_cost + Doctor_conslt_fee + Medicine_bill

print('Total_bill for patient :',Total_Bill)
```

```python
## If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied

if Total_Bill >10000:
    discount = Total_Bill * 0.10
else:
    discount = 0
    
Final_Bill = Total_Bill - discount

print('Final_Bill :',Final_Bill)

print('Doctor_conslt_fee:', Doctor_conslt_fee)
print('Room cost per patient: ',Room_cost)
print('LAB test cost per patient:',Lab_cost)
print ('Total Medicine bill: ',Medicine_bill)
print('Total_Bill : ',Total_Bill)
print('Final_Bill: ',Final_Bill)
```
