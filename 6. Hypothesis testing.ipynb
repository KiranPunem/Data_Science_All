## ESTIMATION AND CONFIDENCE INTERVALS

Scenario:

A manufacturer of print-heads for personal computers is interested in estimating the mean durability 
of their print-heads in terms of the number of characters printed before failure. 
To assess this, the manufacturer conducts a study on a small sample of 
print-heads due to the destructive nature of the testing process.


Data:


A total of 15 print-heads were randomly selected and tested until failure. The durability of each print-head (in millions of characters) was recorded as follows:
1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29


## Assignment Tasks

a. Build 99% Confidence Interval Using Sample Standard Deviation


Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 
                 1.32, 0.85, 1.07, 1.48, 1.20, 1.33,
                 1.18, 1.22, 1.29])

# Statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)  # ddof - Degrees of freedom

# 99% Confidence Interval
confidence = 0.99
alpha = 1-confidence

# t-interval (with out knowing population std)
t_val = stats.t.ppf(1 - alpha/2, n-1)
margin_t = t_val * std / np.sqrt(n)

lower_t = mean - margin_t
upper_t = mean + margin_t

print('99% Confidence Interval Using Sample Standard Deviation (t-interval) :')
print(np.round(lower_t,4))
print(np.round(upper_t,4))
```

```python
# chart for t-interval 

plt.figure(figsize=(4,3))
plt.hist(data, bins=8, density=True)
plt.axvline(mean)
plt.axvline(lower_t, linestyle='--')
plt.axvline(upper_t, linestyle='--')

plt.title('99% Confidence Interval for mean')
plt.xlabel("Durability (million characters)")
plt.ylabel("Density")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(4,3))
sns.histplot(data,bins=8,kde=True)
plt.axvline(mean)
plt.axvline(lower_t, linestyle='--')
plt.axvline(upper_t, linestyle='--')

plt.title('99% Confidence Interval for mean')
plt.xlabel("Durability (million characters)")
plt.ylabel("Density")
plt.show()
```

### b. Build 99% Confidence Interval Using Known Population Standard Deviation

If it were known that the population standard deviation is 0.2 million characters, 
construct a 99% confidence interval for the mean number of characters printed before failure.


```python
#Data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 
                 1.32, 0.85, 1.07, 1.48, 1.20, 1.33,
                 1.18, 1.22, 1.29])

# Statistics
n = len(data)
mean = np.mean(data)
population_std = 0.2

# 99% Confidence Interval
confidence = 0.99
alpha = 1-confidence

# z-interval (with out knowing population std)
z_val = stats.norm.ppf(1 - alpha/2)
margin_z = z_val * ( population_std / np.sqrt(n))

lower_z = mean - margin_z
upper_z = mean + margin_z

print('99% Confidence Interval Using population Standard Deviation (z-interval) :')
print(np.round(lower_z,4))
print(np.round(upper_z,4))
```

```python
# chart for z-interval 

plt.figure(figsize=(4,3))
plt.hist(data, bins=8, density=True)
plt.axvline(mean)
plt.axvline(lower_z, linestyle='--')
plt.axvline(upper_z, linestyle='--')

plt.title('99% Confidence Interval for mean')
plt.xlabel("Durability (million characters)")
plt.ylabel("Density")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(4,3))
sns.histplot(data,bins=8,kde=True)
plt.axvline(mean)
plt.axvline(lower_z, linestyle='--')
plt.axvline(upper_z, linestyle='--')

plt.title('99% Confidence Interval for mean')
plt.xlabel("Durability (million characters)")
plt.ylabel("Density")
plt.show()
```

```python
print('99% Confidence Interval Using Sample Standard Deviation (t-interval) :')
print('Lower_Value :',lower_t)
print('Upper_Value :',upper_t)
print('-'*50)
print('99% Confidence Interval Using population Standard Deviation (z-interval) :')
print('Lower_Value :',lower_z)
print('Upper_Value :',upper_z)
```
