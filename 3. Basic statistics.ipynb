## ●	Box Plots:

	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions.


```python
numeric.plot(kind='box', vert=False, figsize=(10,6))
plt.title("Box Plot of Numeric Variables")
plt.show()

#  More Outliers Observed in Total_Sales, Net_Sales & Discount Amount
# How many Outliers are there, will check individual
```

```python
numeric['Net_Sales_Value'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Net_Sales_Value')
plt.show()
```

```python
# Net_Sales_Value
Q1 = np.percentile(numeric['Net_Sales_Value'],25)
Q3 = np.percentile(numeric['Net_Sales_Value'],75)
IQR = Q3-Q1
# Upper whisker
UW = Q3 + (1.5 * IQR)
print(UW)
Outliers = len(numeric['Net_Sales_Value'][numeric['Net_Sales_Value']>UW])
print('No of Outliers:',Outliers)
```

```python
numeric['Discount_Amount'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Discount_Amount')
plt.show()
```

```python
# Discount_Amount
Q1 = np.percentile(numeric['Discount_Amount'],25)
Q3 = np.percentile(numeric['Discount_Amount'],75)
IQR = Q3-Q1
# Upper whisker
UW = Q3 + (1.5 * IQR)
print(UW)
Outliers = len(numeric['Discount_Amount'][numeric['Discount_Amount']>UW])
print('No of Outliers:',Outliers)
```

```python
numeric['Discount Rate (%)'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Discount Rate (%)')
plt.show()
```

```python
# Discount Rate (%) (Observed LW)
Q1 = np.percentile(numeric['Discount Rate (%)'],25)
Q3 = np.percentile(numeric['Discount Rate (%)'],75)
IQR = Q3-Q1
# Upper whisker
LW = Q1 - (1.5 * IQR)
print(LW)
Outliers = len(numeric['Discount Rate (%)'][numeric['Discount Rate (%)']< LW])
print('No of Outliers:',Outliers)
```

```python
numeric['Total_Sales_Value'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Total_Sales_Value')
plt.show()
```

```python
# Total_Sales_Value
Q1 = np.percentile(numeric['Total_Sales_Value'],25)
Q3 = np.percentile(numeric['Total_Sales_Value'],75)
IQR = Q3-Q1
# Upper whisker
UW = Q3 + (1.5 * IQR)
print(UW)
Outliers = len(numeric['Total_Sales_Value'][numeric['Total_Sales_Value']>UW])
print('No of Outliers:',Outliers)
```

```python
numeric['Avg_Price'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Avg_Price')
plt.show()
```

```python
# Avg_Price (Observed More Outliers)
Q1 = np.percentile(numeric['Avg_Price'],25)
Q3 = np.percentile(numeric['Avg_Price'],75)
IQR = Q3-Q1
# Upper whisker
UW = Q3 + (1.5 * IQR)
print(UW)
Outliers = len(numeric['Avg_Price'][numeric['Avg_Price']>UW])
print('No of Outliers:',Outliers)
```

```python
numeric['Volume'].plot(kind='box', vert=False, figsize=(6,5))
plt.title('Volume')
plt.show()
```

```python
# Volume
Q1 = np.percentile(numeric['Volume'],25)
Q3 = np.percentile(numeric['Volume'],75)
IQR = Q3-Q1
# Upper whisker
UW = Q3 + (1.5 * IQR)
print(UW)
Outliers = len(numeric['Volume'][numeric['Volume']>UW])
print('No of Outliers:',Outliers)
```
