1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
num = [10,20,30,40,50]
print('List Of Integers :',num)
```

```python
#append = used to add another integer

num.append(60)
print('Append :', num)
```

```python
# extend = be can add multiple integers

num.extend([70,80,90])
print('Extend :', num)
```

```python
# insert = be can add specific place with new integer

num.insert(2,25)
print('after insert :', num)
```

```python
# remove = can be a remove value

num.remove(30)
print('after remove :', num)
```

```python
# pop = remove be a element with using index

num.pop(2)
print('after pop :', num)
```

```python
# index = to find integer with using index

num.index(40)
print('index :', num.index(40))
```

```python
# clear = To clear the All integer values

num.clear()
print('after clearing :', num)
```

```python
## sort = to sorting the data/integers

values = [10,15,20,50,70,35,25,65]
print('Before sort :',values)
values.sort()
print('After Sort :', values)
```

```python
## reverse = reversing values

values = [10, 15, 20, 25, 35, 50, 65, 70]
print('Before reverse :',values)
values.reverse()
print('After reverse :', values)

## another way
rev = values[::-1]
print('another way for reverse :', rev)
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
import pandas as pd

data = {'Name':['A_name','B_name','C_name','D_name','E_name','F_name','G_name','X_name','Y_name','Z_name'],
       'Age':[25, 30, 22, 28, 35, 40, 26, 32, 29, 45],
       'City':['SRH','RCB','CSK','MI','DD','KNR','Pune','GL','HYD','BNG'],
       'Salary':[30000, 45000, 28000, 50000, 60000,55000, 32000, 48000, 35000, 65000]}
df = pd.DataFrame(data)
print(df)
```

```python
# use info = to get All info
df.info()
```

```python
# use describe = works numerical only
df.describe()
```

```python
# Select Name and City columns 
df[['Name','City']]
```

```python
# Drop City Column

df.drop(['City'], axis=1)
```

```python
## Fill any missing values in 'Salary' column with the mean

df.loc[[2, 5, 8], 'Salary'] = np.nan
```

```python
df
```

```python
df.info()
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```
